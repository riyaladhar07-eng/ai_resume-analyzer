```python
def analyze_resume(resume_text):
    skills = ["python", "java", "html", "css", "excel", "communication", "machine learning", "ai"]

    score = 0
    missing = []

    resume_text = resume_text.lower()

    for skill in skills:
        if skill in resume_text:
            score += 1
        else:
            missing.append(skill)

    print("\n--- Resume Analysis ---")
    print("Score:", score, "/", len(skills))

    if score >= 6:
        print("Strong Profile 💪")
    elif score >= 3:
        print("Average Profile 🙂")
    else:
        print("Needs Improvement ⚠️")

    print("\nMissing Skills:")
    for skill in missing:
        print("-", skill)


# input
resume = input("Paste your resume text:\n")
analyze_resume(resume)
```
